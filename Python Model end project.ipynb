Step 2: Add More Tickets


```python
next_ticket_num=11
num_tickets=int(input("Enter how many new tickets you want to add?"))
for i in range(num_tickets):
  print(f"/n Entering details for ticket number {next_ticket_num}")
  customer_name=input("Enter customer name: ")
  issue_description=input("Enter issue description: ")

  while True:
        priority=input("Enter ticket priority (High/Medium/Low):").capitalize()
        if priority in ["High","Medium","Low"]:
          break
        else:
          print("Invalid priority! Please enter high, medium or low")

next_ticket_num+=1
ticket_data.append({
        "Ticket No": next_ticket_no,
        "Customer Name": customer_name,
        "Issue_Description": issue_description,
        "Priority": priority
    })


#Display updated ticket details
print("\nUpdated Ticket Data:")
for ticket in ticket_data:
    print(ticket)


```

```python
import pandas as pd
df=pd.DataFrame(ticket_data)
df
```

```python
df["Issue_Description"]=df["Issue_Description"].replace("okay","ok")
df
```

Step 3: Text Cleaning for Issue Descriptions

```python
df["Issue_Description"]=df["Issue_Description"].str.lower()
df["Issue_Description"]=df["Issue_Description"].str.strip()
df["Issue_Description"]=df["Issue_Description"].str.replace("!","")
df["Issue_Description"]=df["Issue_Description"].str.replace(".","")
df
```

 Step 4: Keyword-Based Issue Insights

```python
def count_tickets_with_word(word):
  count=0
  word= word.lower()

  for desc in df["Issue_Description"]:
    if word in desc:
      count+=1
  return count
```

```python
print("Tickets containing 'poor':", count_tickets_with_word("poor"))
print("Tickets containing 'good':", count_tickets_with_word("good"))
print("Tickets containing 'slow':", count_tickets_with_word("slow"))
print("Tickets containing 'excellent':", count_tickets_with_word("excellent"))
```

3.	Find the Ticket With the Longest Issue Description
Based on word count, print:
●	Ticket number
●	Customer name
●	Cleaned issue text
●	Word count


```python
longest_ticket = df.assign(
    Word_Count=df["Issue_Description"].str.split().str.len()
).loc[lambda x: x["Word_Count"].idxmax()]

print("\nTicket With Longest Issue Description:\n")

print("Ticket Number:", longest_ticket["Ticket_No"])
print("Customer Name:", longest_ticket["Customer_Name"])
print("Cleaned Issue Text:", longest_ticket["Issue_Description"])
print("Word Count:", longest_ticket["Word_Count"])
```

4.	Extract Unique Words Used
Generate a set of all unique words across all issue descriptions.
Show:
●	Count of unique words
●	The word list (sorted)



```python
# Combine all issue descriptions into one string
all_text = " ".join(df["Issue_Description"].astype(str))

# Convert to lowercase and split into words
words = all_text.lower().split()

# Create a set of unique words
unique_words = set(words)

# Sort the words
sorted_unique_words = sorted(unique_words)

print("Count of unique words:", len(sorted_unique_words))
print("\nList of unique words:")
print(sorted_unique_words)
```
